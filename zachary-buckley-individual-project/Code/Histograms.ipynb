```python
# nice function aaron found. awesome.
def human(num, units = 'M'):
    units = units.lower()
    num = float(num)
    if units == 'k':
        return str(num/10**3) + " K"
    elif units == 'm':
        return str(num/10**6) + " M"
    elif units == 'b':
        return str(num/10**9) +  " B"
```

```python
# copied and slightly modified from aaron's code
fig, ax = plt.subplots()
plt.title('Distribution of Ratings (all)', fontsize=15)
sns.countplot(df['rating'])
ax.set_yticklabels([human(item, 'M') for item in ax.get_yticks()])
ax.set_ylabel('Number of Ratings (Millions)')

plt.show()
```

```python
# copied and slightly modified from aaron's code
fig, ax = plt.subplots()
plt.title('Distribution of Ratings (sample)', fontsize=15)
sns.countplot(['rating'])
ax.set_yticklabels([human(item, 'M') for item in ax.get_yticks()])
ax.set_ylabel('Number of Ratings (Millions)')

plt.show()
```
